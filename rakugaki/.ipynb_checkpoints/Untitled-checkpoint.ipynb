```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn import linear_model
from sklearn import svm
```

```python
import urllib.request as req
url = 'https://raw.githubusercontent.com/maskot1977/ipython_notebook/master/toydata/winequality-red.txt'
req.urlretrieve(url, 'red-wine.txt')
```

```python
df1 = pd.read_csv('red-wine.txt', sep='\t', index_col=0) 
```

```python
df1.head()
```

```python
x_data = df1.drop('quality', axis=1)
y = df1['quality']
```

```python
sns.pairplot(x_data)
```

```python
x_data.head()
```

```python
x = x_data.copy()
```

```python
for key in x:
    if x[key].skew() > 0.75:
        x[key] = np.log(x[key] + 1)
```

```python
x.head()
```

```python
sns.pairplot(x)
```

```python
x['alcohol'].skew()
```

```python
for key1 in x_data:
    x[key1 + '05'] = np.sqrt(x[key1])
    x[key1 + '3'] = x[key1] ** 3
    for key2 in x_data:
        x[key1 + key2] = x[key1] * x[key2]
        x[key1 + 2 * key2] = x[key1] * x[key2] ** 2
        x[key1 * 2 + key2] = x[key1] ** 2 * x[key2] 
```

```python
x.head()
```

```python
x = (x - x.mean(axis=0))/x.std(axis=0)
```

```python
x_train = x[:1200]
x_val = x[1200:]
y_train = y[:1200]
y_val = y[1200:]
```

```python
model1 = svm.SVC(kernel='linear')
model1.fit(x_train, y_train)
```

```python
model1.score(x_train, y_train)
```

```python
model1.score(x_val, y_val)
```

```python
model2 = linear_model.LinearRegression()
model2.fit(x,y)
model2.score()
```
