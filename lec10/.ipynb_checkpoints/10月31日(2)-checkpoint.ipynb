```python
import pandas as pd
df = pd.read_csv('icecream_chosa.txt', sep='\s+', index_col=0)
matrix = df.iloc[:,4:]
```

```python
val_names = matrix.columns.tolist()
```

```python
# リストの正規化をする（Z値に変換する）関数
def normalize(data):
    data = (data-data.mean())/data.std()
```

```python
# リストのリストの正規化をする（Z値に変換する）関数
def normalize2(data):
    list = []
    for x in data:
        list.append(normalize(x))
    return list
```

因子分析を行う以下のコードを実行してください。

```python
# 図やグラフを図示するためのライブラリをインポートする。
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 機械学習のライブラリから、因子分析のプログラムをインポート
from sklearn.decomposition import FactorAnalysis
```

```python
matrix.head()
```

```python
# 因子分析の実行
fa = FactorAnalysis()
factors = fa.fit_transform(matrix.T)
```

```python
# 左から順に第一因子、第二因子、、、
pd.DataFrame(factors, index=val_names)
```

```python
# 第一因子と第二因子でプロットする
plt.figure(figsize=(8, 8))
for x, y, name in zip(factors[:, 0], factors[:, 1], val_names):
    plt.text(x, y, name, alpha=0.8, size=10)
plt.scatter(factors[:, 0], factors[:, 1], alpha=0.8)
plt.title("Factor Analysis")
plt.xlabel("The first factor score")
plt.ylabel("The second factor score")
plt.grid(True)
plt.show()
```

<h4 style="padding: 0.25em 0.5em;color: #494949;background: transparent;border-left: solid 5px #7db4e6;">課題8.2</h4>

上と同様にして、第三因子と第四因子でプロットしてください。また、以上の結果から、アイスクリームの味の好みについて何が言えそうか考察してください。

```python
# 第一因子と第二因子でプロットする
plt.figure(figsize=(8, 8))
for x, y, name in zip(factors[:, 3], factors[:, 4], val_names):
    plt.text(x, y, name, alpha=0.8, size=10)
plt.scatter(factors[:, 3], factors[:, 4], alpha=0.8)
plt.title("Factor Analysis")
plt.xlabel("The third factor score")
plt.ylabel("The fourth factor score")
plt.grid(True)
plt.show()
```
